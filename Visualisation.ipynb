```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/online_shoppers_intention.csv')
```

# ProductRelated_Duration Histogram

```python
plt.figure(figsize=(6, 6))
plt.hist(data['ProductRelated_Duration'], bins=30, color='skyblue', edgecolor='black')
plt.title('ProductRelated_Duration Histogram')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# ProductRelated_Duration Box Plot

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data['ProductRelated_Duration'], vert=False)
plt.title('ProductRelated_Duration Box Plot')
plt.xlabel('Duration')
plt.grid(True)
plt.savefig('product_related_duration_box_plot.png')
plt.show()
```

# ProductRelated_Duration Distribution (Kernel Density Plot)

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data['ProductRelated_Duration'], color='skyblue', fill=True)
plt.title('ProductRelated_Duration Distribution (Kernel Density Plot)')
plt.xlabel('Duration')
plt.ylabel('Density')
plt.grid(True)
plt.savefig('product_related_duration_plot.png')
plt.show()
```

# Kernel Density Plots For BounceRates, ExitRates, PageValues

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# kernel density plot for BounceRates
sns.kdeplot(data['BounceRates'], color='skyblue', fill=True, ax=axes[0])
axes[0].set_title('BounceRates Distribution (Kernel Density Plot)')
axes[0].set_xlabel('BounceRates')
axes[0].set_ylabel('Density')
plt.savefig('BounceRates_kernel_density_plot.png')

# kernel density plot for ExitRates
sns.kdeplot(data['ExitRates'], color='green', fill=True, ax=axes[1])
axes[1].set_title('ExitRates Distribution (Kernel Density Plot)')
axes[1].set_xlabel('ExitRates')
axes[1].set_ylabel('Density')
plt.savefig('ExitRates_kernel_density_plot.png')

# kernel density plot for PageValues
sns.kdeplot(data['PageValues'], color='orange', fill=True, ax=axes[2])
axes[2].set_title('PageValues Distribution (Kernel Density Plot)')
axes[2].set_xlabel('PageValues')
axes[2].set_ylabel('Density')
plt.savefig('PageValues_kernel_density_plot.png')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

# Heatmap For BounceRates, ExitRates, PageValues

```python
correlation_matrix = data[['BounceRates', 'ExitRates', 'PageValues']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap (BounceRates, ExitRates, PageValues)')
plt.savefig('correlation_matrix.png')
plt.show()
```

# PageValues Over Months

```python
month_order = ["Feb", "Mar", "Apr", "May", "June", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Month', y='PageValues', errorbar=None, sort=False, hue=None, hue_order=None)
plt.title('PageValues Over Months')
plt.xlabel('Month')
plt.ylabel('PageValues')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.savefig('pageValues_over_Months.png')
plt.show()
```
